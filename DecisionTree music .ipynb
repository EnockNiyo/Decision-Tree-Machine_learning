```python
import numpy as np      # linear algebra 
import pandas as pd   # data processing, CSV file I/O (e.g. pd.read_csv) 
import matplotlib.pyplot as plt     # data visualization 
from sklearn.tree import DecisionTreeClassifier 
from  sklearn.model_selection import train_test_split #used to split into train model and test
from sklearn.metrics import accuracy_score  # used to calcurate accurancy
#from sklearn.externals import joblib   # joblib object has method from saving and loading models
from sklearn import tree
import seaborn as sns         # statistical data visualizatio
music_data = pd.read_csv('music.csv')     #Read csv file called car_evaluation.csv // is our dataset 
music_data.describe()     # displaying description of your data 
music_data.head()      # shows 
#Preparing the Data 
X = music_data.drop(columns=['genre'])  
y = music_data['genre'] 
X_train, X_test, n, y_test = train_test_split(X, y ,test_size=0.2) # X_train and X_test are INPUT SET , then y_train and y_test are output set 

# colling Model / class DecisionTreeClassifier
model = DecisionTreeClassifier()  
#model.fit(X, y) # X for input and y for predicted data
model.fit(X_train, y_train) # X for input and y for predicted data
predictions = model.predict(X_test)   # method used to make pr
#predictions = model.predict() 
tree.plot_tree(model)
score = accuracy_score(y_test, predictions)
score 
```
